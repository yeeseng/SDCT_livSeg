```python
plt.figure(figsize = (16,40))

numRows = 10

for i in range(numRows):
    randomFile = random.choice(fileList)
    print(randomFile)

    combinedImage = np.load('pyOsiriXimageExports/combined/'+randomFile)
    mask = np.load('pyOsiriXimageExports/mask/'+randomFile)
    comptonImage = np.load('pyOsiriXimageExports/compton/'+randomFile)
    peImage = np.load('pyOsiriXimageExports/pe/'+randomFile)

    plt.subplot(numRows,4,i*4+1)
    plt.imshow(combinedImage.T, cmap='gray')
    
    
    plt.subplot(numRows,4,i*4+2)
    plt.imshow(mask.T, cmap='gray')
    
    plt.subplot(numRows,4,i*4+3)
    plt.imshow(comptonImage.T, cmap='gray')
    
    plt.subplot(numRows,4,i*4+4)
    plt.imshow(peImage.T, cmap='gray')
```
